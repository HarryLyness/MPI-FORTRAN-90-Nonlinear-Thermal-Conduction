```python
import numpy as np
from matplotlib import pyplot as plt
dataq7a = np.loadtxt('dataq9.txt')
m32 = dataq7a[0]
m128=dataq7a[1]
m512=dataq7a[2]

P = np.array([1,2,4,8,16,32],dtype = float)
plt.plot(P,m32 , marker=".", markersize=8, label = 'm=32')
plt.plot(P,m128 , marker=".", markersize=8, label = 'm=128')
plt.plot(P,m512 , marker=".", markersize=8, label = 'm=512')
plt.xlabel('Number of Processors P')
plt.ylabel('time (s)')
plt.title(' Average time per CG iteration for m')
plt.legend()
plt.show()
#plt.savefig('plot_timeq9.pdf')

plt.plot(P,m32[0]/(P*m32), marker=".", markersize=8, label = 'm=32')
plt.plot(P,m128[0]/(P*m128), marker=".", markersize=8, label = 'm=128')
plt.plot(P,m512[0]/(P*m512), marker=".", markersize=8, label = 'm=512')
plt.xlabel('Number of Processors P')
plt.ylabel('Parallel Efficiency E(n,P)')
plt.title(' Parallel Efficiency measured per CG iteration for n=m^2')
plt.legend()
plt.show()
#plt.savefig('plot_effq9.pdf')

plt.plot(P,m32[0]/(m32), marker=".", markersize=8, label = 'm=32')
plt.plot(P,m128[0]/(m128), marker=".", markersize=8, label = 'm=128')
plt.plot(P,m512[0]/(m512), marker=".", markersize=8, label = 'm=512')
plt.xlabel('Number of Processors P')
plt.ylabel('Speed up S(n,P)')
plt.title(' Parallel Speed up measured per CG iteration for n=m^2')
plt.legend()
plt.show()

# creating a nice table :)
tpar = np.array([P,tpar32,m32,tpar128,m128,tpar512,m512])
tpar = tpar.T
import pandas as pd
tpard = pd.DataFrame(tpar)
tpard.columns = ['Number of Processors', 'Pred. time (s) m=32','Meas. time (s) m=32', 'Pred. time (s) m=128','Meas. time (s) m=128', 'Pred. time (s) m=512','Meas. time (s) m=512']
#tpard.rows = ['1','2','4','8','16','32']
tpard
```
